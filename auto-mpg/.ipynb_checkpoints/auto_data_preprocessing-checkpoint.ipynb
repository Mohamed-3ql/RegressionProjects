```python
# import libraries required for data analysis and visualization
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


from IPython.display import display

%matplotlib inline
```

#### read data and check it

```python
autos_df = pd.read_csv("auto-mpg.data-original",delim_whitespace=True,header = None)
autos_df.head()
```

```python
# set header_attributes for your dataframe
header = ['mpg','cylinders','displacement','horsepower','weight',
         'accelration','model_year','origin','car_name']
```

```python
# set header to your dataframe
autos_df.columns = header
```

```python
#check our dataframe now to make sure everything is going right
autos_df.head(50)
```

```python
# check data dimensions
autos_df.shape
```

```python
# get basic info about dataset
autos_df.info()
```

#### working with missing data
* identify missing data
* deal with missing data
* correct data format

```python
# identifing missing values
missing_data = autos_df.isnull()
# itertate through whole set of data to count true and false values
for col in missing_data.columns.values.tolist():
    print(col)
    print(missing_data[col].value_counts())
```

#### dealing with missing value
* replace horsepower missing values with the mean of horsepower
* drop the data of label y (dependent variable) ("mpg")

```python
# calculate the mean then replace it with nan values in horsepower column
horsepow_mean = autos_df["horsepower"].astype("float").mean(axis=0)
autos_df["horsepower"].replace(np.nan,horsepow_mean,inplace=True)
```

```python
# check out dataset
autos_df.info()
```

```python
# we get some info from previous operation that we have null values(missing values)
# these missing values in "mpg" column
# since mpg is our y that we want to predicate.we should drop the whole record 
autos_df.dropna(subset=['mpg'],axis=0,inplace=True)
autos_df.reset_index(drop=True,inplace=True)
```

```python
#after drop the row in "mpg" missing values
# check out our dataset again
autos_df.info()
```

```python
# check our dataset to see if we need to correct somedat format or not
autos_df.dtypes
```

#### correct data format
* cylinder feature is multi-value discrete variable
* model_year (multi-value discrete)
* origin (multi-value discrete)

```python
# convert each of "cylinders","model_year" and "origin" to integer to correct data format
autos_df[['cylinders', 'model_year','origin']]= autos_df[['cylinders', 'model_year', 'origin']].astype("int")
```

```python
# check out the dataset to make sure from the changes that we made
autos_df.dtypes
```

### Data Standardization

* example 
    * transform "mpg" to L/100km

```python
# adding new feature by transform "mpg" to L/100km
autos_df["L/100"] = 235/autos_df["mpg"]
```

```python
# check out the dataframe
autos_df.head()
```

```python
# get max and min value of each continous variable
cont_var_features = ["displacement","horsepower","weight","accelration"]
i = 1
for col in cont_var_features:
    print(str(i)+"-"+col)
    print("Minimum value: {0} and Maximum Value: {1}".
          format(min(autos_df[col]),max(autos_df[col])))
    print()
    i = i +1
```

```python
#simple normalization for weight
autos_df['weight'] = autos_df['weight'].astype(int,copy=True)
# normalize weight by devide weight by max(weight)
autos_df['weight'] = autos_df['weight']/autos_df['weight'].max()
```

```python
# check out "weight" column after we did weight normalization
print(autos_df['weight'].min())
print(autos_df['weight'].max())
```

```python
# binning both of "horsepower" and "displacement" columns
# convert "displacement" and "horsepower" into integer type
print("Displacement Minimum Value: ",autos_df['displacement'].min())
print("Displacement Maximum Value: ",autos_df['displacement'].max())
print("Horsepower Minimum Value: ",autos_df['horsepower'].min())
print("Horsepower Maximum Value: ",autos_df['horsepower'].max())
# convert type of displacement and horsepower from float' to int
autos_df["displacement"] = autos_df['displacement'].astype(int,copy=True)
autos_df["horsepower"] = autos_df['horsepower'].astype(int, copy=True)

print("Displacement Minimum Value: ",autos_df['displacement'].min())
print("Displacement Maximum Value: ",autos_df['displacement'].max())
print("Horsepower Minimum Value: ",autos_df['horsepower'].min())
print("Horsepower Maximum Value: ",autos_df['horsepower'].max())
```

```python
# display hisotogram for  "displacement" 
plt.hist(autos_df['displacement'])
plt.xlabel("Displacement")
plt.ylabel("Frequency")
plt.show()
```

```python
# display hisotogram for  "horsepower"
plt.hist(autos_df["horsepower"])
plt.xlabel("horsepower")
plt.ylabel("Frequency")
plt.show()
```

```python
#create bins for both of features displacement and horsepower
horse_bins = np.linspace(min(autos_df['horsepower']),max(autos_df['horsepower']),4)
disp_bins = np.linspace(min(autos_df['displacement']),max(autos_df['displacement']),4)
group_names = ["low","medium","high"]

autos_df["horsepower_binned"] = pd.cut(autos_df["horsepower"],horse_bins,labels=group_names,include_lowest=True)
autos_df["displacement_binned"] = pd.cut(autos_df["displacement"],disp_bins,labels=group_names,include_lowest=True)
# check the result
autos_df[["horsepower","horsepower_binned","displacement","displacement_binned"]].head()
```

```python
# let's see the number of vehicles in displacement_binned
autos_df["displacement_binned"].value_counts()
```

```python
# let's see the number of vehicles in  horsepower_binned
autos_df["horsepower_binned"].value_counts()
```

```python
# diplay the bar graph horsepower_binned
plt.bar(group_names,autos_df["horsepower_binned"].value_counts())
plt.xlabel("Horsepower bins")
plt.ylabel("Frequency")
plt.title("horsepower distrbution")
plt.show()
```

```python
# display the bar graph displacement_binned
plt.bar(group_names,autos_df["displacement_binned"].value_counts())
plt.xlabel("displacement bins")
plt.ylabel("Frequency")
plt.title("Displacement distrbution")
plt.show()
```
